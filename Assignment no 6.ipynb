### Question No 1
#### ANSWER:
Object-oriented programming (OOP) is a way of writing computer programs which is using the idea of "objects" to represent data and methods. Usually, computer programs were just a list of instructions to the computer, telling the computer to do certain things in a certain way, which is called procedural programming. However, in object-oriented programming, computer programs make use of objects that talk to one another and to change the data in those objects, to work in a way that the user wants. Also, because of the way object-oriented programming is designed, it helps the developer by allowing for code to be easily reused by other parts of the program or even by other people.

### Question No 4
#### ANSWER:
###### CLASS: A class is an entity that determines how an object will behave and what the object will contain. In other words, it is a blueprint or a set of instruction to build a specific type of object.
###### OBJECT:An object is nothing but a self-contained component which consists of methods and properties to make a particular type of data useful.
###### BEHAVIOUR:The behavior of an object is defined by its methods.Methods determine what type of functionality a class has, how it modifies its data, and its overall behavior.
###### STATE:The states of object are the instances inside the object that contains the data.

```python
#QUESTION NO 5
class Car():
    def __init__(self,model,color,name,price,condition):
        self.model=model
        self.color=color
        self.name=name
        self.price=price
        self.condition=condition
        self.companyname='Suzuki'
        self.tax=4000
        self.maintanance=500
    def updatecomp_name(self,a):
        self.companyname=a
    def updatetax_price(self,b):
        self.tax=b
    def updatemaintanance_price(self,c):
        self.maintanance=c

c1=Car(17,'Red',"Civic",500000,"new")
print(c1.model)
print(c1.color)
print(c1.name)
print(c1.price)
print(c1.condition)
print("Old company name: "+c1.companyname)
c1.updatecomp_name("Azhar")
print("New company name: "+c1.companyname)
print("Old Tax rate: ",c1.tax)
c1.updatetax_price(5000)
print("New Tax rate: ",c1.tax)
print("Old Maintanance rate: ",c1.maintanance)
c1.updatemaintanance_price(1000)
print("New Maintanance rate: ",c1.maintanance)
print()

c2=Car(18,'White',"Carola",100000,"use")
print(c2.model)
print(c2.color)
print(c2.name)
print(c2.price)
print(c2.condition)
print("Old company name: "+c2.companyname)
c2.updatecomp_name("VEHICLES")
print("New company name: "+c2.companyname)
print("Old Tax rate: ",c2.tax)
c2.updatetax_price(6000)
print("New Tax rate: ",c2.tax)
print("Old Maintanance rate: ",c2.maintanance)
c2.updatemaintanance_price(1200)
print("New Maintanance rate: ",c2.maintanance)
print()

c3=Car(19,'Blue',"Cultus",10000000,"new")
print(c3.model)
print(c3.color)
print(c3.name)
print(c3.price)
print(c3.condition)
print("Old company name: "+c3.companyname)
c3.updatecomp_name("TAYOTA")
print("New company name: "+c3.companyname)
print("Old Tax rate: ",c3.tax)
c3.updatetax_price(5200)
print("New Tax rate: ",c3.tax)
print("Old Maintanance rate: ",c3.maintanance)
c3.updatemaintanance_price(1100)
print("New Maintanance rate: ",c3.maintanance)
print()

c4=Car(15,'Red',"Vitz",500000,"new")
print(c4.model)
print(c4.color)
print(c4.name)
print(c4.price)
print(c4.condition)
print("Old company name: "+c4.companyname)
c4.updatecomp_name("HWAVI")
print("New company name: "+c4.companyname)
print("Old Tax rate: ",c4.tax)
c4.updatetax_price(5000)
print("New Tax rate: ",c4.tax)
print("Old Maintanance rate: ",c4.maintanance)
c4.updatemaintanance_price(1600)
print("New Maintanance rate: ",c4.maintanance)
print()

c5=Car(17,'Black',"Mazda",500000,"new")
print(c5.model)
print(c5.color)
print(c5.name)
print(c5.price)
print(c5.condition)
print("Old company name: "+c5.companyname)
c5.updatecomp_name("AUDI")
print("New company name: "+c5.companyname)
print("Old Tax rate: ",c5.tax)
c5.updatetax_price(7000)
print("New Tax rate: ",c5.tax)
print("Old Maintanance rate: ",c5.maintanance)
c5.updatemaintanance_price(2000)
print("New Maintanance rate: ",c5.maintanance)
print()
```
